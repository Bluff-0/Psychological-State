```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

# plt.rcParams['figure.figsize'] = 5,5
```

```python
df= pd.read_csv('Predictions/Predictions_CNN_MLP_Male_TESS.csv')
sns.heatmap(confusion_matrix(np.array(df.iloc[:,0]), np.array(df.iloc[:,1])), annot=True)
plt.title('TESS OLD CONFUSUION MATRIX')
plt.show()
```

```python
print("\t\tTESS OLD CLASSIFICATION REPORT\n")
print(classification_report(np.array(df.iloc[:,0]), np.array(df.iloc[:,1])))
```
